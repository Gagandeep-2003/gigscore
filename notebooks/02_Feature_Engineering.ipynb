# 🔧 GigScore — Feature Engineering

This notebook demonstrates the feature engineering pipeline:
1. Traditional credit features (income ratios, credit terms)
2. Gig-specific features (platform reliability, income stability)
3. India-specific features (UPI engagement, settlement stability)
4. Feature pipeline construction (sklearn Pipeline)

```python
import pandas as pd
import numpy as np
import sys
sys.path.insert(0, '..')

from src.data.synthetic_generator import generate_synthetic_gig_data
from src.features.traditional_features import engineer_traditional_features
from src.features.gig_features import engineer_gig_features
from src.features.india_features import engineer_india_features
from src.features.feature_pipeline import build_feature_pipeline, get_available_features

# Generate data
df = generate_synthetic_gig_data(n_samples=10000, seed=42)
print(f'Starting columns: {len(df.columns)}')
```

```python
# Apply feature engineering
df = engineer_traditional_features(df)
df = engineer_gig_features(df)
df = engineer_india_features(df)
print(f'After engineering: {len(df.columns)} columns')
print(f'\nNew features:')
new_feats = ['income_stability_score', 'platform_reliability_score', 
             'financial_resilience_score', 'digital_engagement_score',
             'settlement_stability_score', 'lifestyle_risk_score',
             'multi_platform_bonus', 'work_intensity_score']
for f in new_feats:
    if f in df.columns:
        print(f'  {f}: mean={df[f].mean():.3f}, std={df[f].std():.3f}')
```
